```python
import pandas as pd
import plotly.express as px


df = pd.read_csv("data/most_subscribed_youtube_channels.csv")

df["subscribers"] = ( df["subscribers"].astype(str).str.replace(",", "") .astype(int)
)


print(df.head())


fig_hist = px.histogram(
    df,
    x="subscribers",
    title="Distribution of YouTube Channel Subscribers",
    labels={"subscribers": "Number of Subscribers"},
    nbins=20
)
fig_hist.show()

top_10 = df.sort_values("subscribers", ascending=False).head(10)

fig_pie = px.pie(
    top_10,
    names="channel_name",
    values="subscribers",
    title="Top 10 Most Subscribed YouTube Channels"
)
fig_pie.show()

fig_box = px.box(
    df,
    y="subscribers",
    title="Subscriber Count Distribution (Box Plot)"
)
fig_box.show()
```
